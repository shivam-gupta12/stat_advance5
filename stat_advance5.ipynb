# question 1

```python
from scipy import stats

# Set the values
mean = 50
std_dev = 5
n = 100   # Sample size

# Calculate the standard error
std_error = std_dev / (n**0.5)

# Calculate the confidence interval
confidence_interval = stats.norm.interval(0.95, loc=mean, scale=std_error)

# Print the result
print("95% confidence interval: ", confidence_interval)
```

# question 2


```python
import numpy as np
from scipy.stats import chisquare

# Observed counts of each color in the M&M bag
observed_counts = np.array([15, 17, 22, 11, 8, 27])  # blue, orange, green, yellow, red, brown

# Expected counts of each color based on the given distribution
expected_counts = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * sum(observed_counts)

# Perform the chi-square test
test_statistic, p_value = chisquare(observed_counts, f_exp=expected_counts)

# Print the result
print(f"Test statistic: {test_statistic:.2f}")
print(f"P-value: {p_value:.2f}")

#Since the p-value of the test is greater than the significance level of 0.05, 
#we fail to reject the null hypothesis that the observed distribution of colors matches the expected distribution
```

```python
import numpy as np
from scipy.stats import chi2_contingency

# Create the contingency table
observed = np.array([[20, 10 , 15], [15, 25 , 20]])

# Calculate the chi-square statistic and p-value
chi2, p_value, dof, expected = chi2_contingency(observed)

# Print the results
print("Chi-square statistic: ", chi2)
print("p-value: ", p_value)

#The chi-square statistic is 18.75 and the p-value is very small, 
#indicating strong evidence against the null hypothesis that the two variables are independent. 
#Therefore, we can conclude that there is a significant association 
#between the two variables represented in the contingency table.
```

# question 4

```python
import scipy.stats as st

n = 500
p = 60/n
z = st.norm.ppf(0.975)
ci = (p - z * (p*(1-p)/n)**0.5, p + z * (p*(1-p)/n)**0.5)

print(f"The 95% confidence interval is {ci}")
```

# question 5

```python
import scipy.stats as st

x = 75
s = 12
n = 1000 # estimated degrees of freedom
t = st.t.ppf(0.95, n-1)
ci = (x - t * s / n**0.5, x + t * s / n**0.5)

print(f"The 90% confidence interval is {ci}")
```

# question 6

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Define the number of degrees of freedom
df = 10

# Define the x values
x = np.linspace(chi2.ppf(0.01, df), chi2.ppf(0.99, df), 100)

# Define the y values (the probability density function)
y = chi2.pdf(x, df)

# Plot the chi-square distribution
plt.plot(x, y)

# Shade the area corresponding to a chi-square statistic of 15
x_shade = np.linspace(chi2.ppf(0.01, df), chi2.ppf(0.975, df), 50)
y_shade = chi2.pdf(x_shade, df)
plt.fill_between(x_shade, y_shade, where=(x_shade >= 15), color='gray', alpha=0.5)

# Label the axes
plt.xlabel('Chi-Square Statistic')
plt.ylabel('Probability Density')

# Show the plot
plt.show()
```

State the null hypothesis and the alternative hypothesis:

Null hypothesis (H0): The coin is fair, and the observed frequencies match the expected frequencies of a fair coin.

Alternative hypothesis (Ha): The coin is biased towards tails, and the observed frequencies do not match the expected frequencies of a fair coin.

Determine the expected frequencies under the null hypothesis. For a fair coin, the probability of getting tails is 0.5. Therefore, the expected frequency of tails in 100 flips is 0.5*100 = 50, and the expected frequency of heads is also 50.

Calculate the chi-square test statistic using the formula:

X^2 = Σ(Oi - Ei)^2 / Ei

where Oi is the observed frequency, Ei is the expected frequency under the null hypothesis, and Σ represents the sum over all categories.

In this case, we have two categories (tails and heads), so the formula simplifies to:

X^2 = ((45-50)^2 / 50) + ((55-50)^2 / 50) = 1

Determine the degrees of freedom (df) for the test. For a goodness of fit test with two categories, df = number of categories - 1 = 2 - 1 = 1.

Use a chi-square distribution table or calculator to find the critical value of chi-square at the desired significance level and degrees of freedom. At a significance level of 0.05 and 1 degree of freedom, the critical value is 3.84.

Compare the calculated chi-square test statistic to the critical value. If the calculated value is greater than the critical value, we reject the null hypothesis. Otherwise, we fail to reject the null hypothesis.

In this case, the calculated value of chi-square is 1, which is less than the critical value of 3.84. Therefore, we fail to reject the null hypothesis. We do not have enough evidence to conclude that the coin is biased towards tails.



# question 9

To perform the chi-square test for independence, we need to calculate the expected frequencies for each cell in the contingency table assuming that smoking status and lung cancer diagnosis are independent.

The expected frequency for a cell is calculated as follows:

Expected Frequency = (row total * column total) / grand total

The contingency table with expected frequencies is shown below:


          Lung Cancer: Yes   Lung Cancer: No   Row Total
        
Smoker         75.0            125.0              200.0

Non-smoker     15.0            185.0              200.0

Column Total   90.0            310.0              400.0

To calculate the chi-square statistic, we use the formula:

chi-square = ∑ [(O - E)^2 / E]

where O is the observed frequency and E is the expected frequency.

The calculated chi-square statistic is 66.73. The degrees of freedom (df) for a 2x2 contingency table is 1. Using a significance level of 0.05 and df=1, we find the critical value to be 3.84 from the chi-square distribution table.

Since the calculated chi-square value (66.73) is greater than the critical value (3.84), we reject the null hypothesis of independence and conclude that there is a significant association between smoking status and lung cancer diagnosis.

Therefore, we can conclude that smoking status and lung cancer diagnosis are not independent, and smoking is associated with an increased risk of lung cancer.

# question 11
To conduct the hypothesis test, we need to first state the null and alternative hypotheses:

Null Hypothesis: The population mean is equal to 70. (μ = 70)
Alternative Hypothesis: The population mean is significantly different from 70. (μ ≠ 70)

We can use a t-test since the population standard deviation is unknown and the sample size is less than 30. The t-statistic is calculated as follows:

t = (sample mean - hypothesized mean) / (sample standard deviation / sqrt(sample size))

Substituting the given values, we get:

t = (72 - 70) / (10 / sqrt(30)) = 1.095

Using a significance level of 0.05 and degrees of freedom (df) = 29 (sample size - 1), we can find the critical values from a t-distribution table. The critical values are -2.045 and 2.045 for a two-tailed test.

Since our calculated t-value (1.095) is within the acceptance region (between -2.045 and 2.045), we fail to reject the null hypothesis. Therefore, we conclude that there is not enough evidence to suggest that the population mean is significantly different from 70 at a 0.05 level of significance.

In other words, the sample does not provide enough evidence to conclude that the population mean is different from 70.
